1.4 We know that standard deviation is a measure of spread in the dataset. What is meant by deviation here? What is the formula for calculating deviation? Can deviation be negative?

- Deviation is the result of substracting the mean from each data point. The formula is (X[i]-mean(X) were X[i] is the data point, for all i values in the list X, or (x-mean(x)) if x is the variable that takes the value of each item in the list.

- Yes, it can be negative:
 X = [1,2,3] ; mean(X) = 3; X(0)-mean(X)


```python
import statistics as stats
X = [1,2,3]
stats.mean(X)
X[0]-stats.mean(X)
```
